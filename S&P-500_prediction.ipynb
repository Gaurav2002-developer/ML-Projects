```python
import yfinance as yf
import pandas as pd
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
df = sp500.history(period = 'max')
```

```python
df.drop(columns = ['Dividends' , 'Stock Splits'] , inplace =True)
```

```python
df.head()
```

```python
df['Tomorrow'] = df['Close'].shift(-1)
```

```python
df.head()
```

```python
df['Target'] = (df['Tomorrow'] > df['Close']).astype(int)
```

```python
df = df.loc['1990-01-01' :].copy()
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 100 , min_samples_split = 100,
                               random_state = 1)

```

```python
train = df.iloc[:-100]
test = df.iloc[-100:]

predictors = ['Close' , 'High' , 'Low' , 'Volume' , 'Low']
model.fit(train[predictors], train['Target'])
```

```python
from sklearn.metrics import precision_score
```

```python
preds = model.predict(test[predictors])
```

```python
def predict(train, test, predictors,  model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index , name = 'Predictions')
    combined = pd.concat([test['Target'], preds], axis = 1)
    return combined
```

```python
def backtest(data, model, predictors, start = 2500 , step = 250):
    all_predictors = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i :(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictors.append(predictions)

    return pd.concat(all_predictors)
```

```python
predictions = backtest(df , model, predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
predictions['Target'].value_counts() / predictions.shape[0]
```

```python
horizon = [2,5,60,250, 1000]

new_predictors = []
for h in horizon:
    rolling_averages = df.rolling(h).mean()

    ratio_column = f"Close_Ratio_{h}"
    df[ratio_column] = df['Close'] / rolling_averages['Close']

    trend_column = f'Trend_{h}'
    df[trend_column] = df.shift(1).rolling(h).sum()['Target']

    new_predictors += [ratio_column, trend_column]
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 50, random_state =1 )
```

```python
def predict(train, test, predictors,  model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index = test.index , name = 'Predictions')
    combined = pd.concat([test['Target'], preds], axis = 1)
    return combined
```

```python
predictions = backtest(df,  model, new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```
