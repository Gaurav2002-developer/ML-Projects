```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from  sklearn.linear_model import LogisticRegression

print("Imports successful!")
```

```python
diabetes = pd.read_csv("diabetes.csv")
diabetes.head()
```

```python
diabetes.shape
```

```python
diabetes.info()
```

```python
diabetes['Outcome'].value_counts()
```

```python
X = diabetes.drop('Outcome',axis=1)
y = diabetes['Outcome']
```

```python
X.shape
```

```python
y.shape
```

```python
scalr = StandardScaler()
scalr.fit(X)
starded_data = scalr.fit_transform(X)
X = starded_data
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state=1)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
svc = svm.SVC(kernel = "linear")

svc.fit(X_train,y_train)
```

```python
train_y_pred = svc.predict(X_train)
test_y_pred = svc.predict(X_test)
```

```python
print("Train set accuracy:", accuracy_score(y_train, train_y_pred))
print("Test set accuracy:", accuracy_score(y_test, test_y_pred))
```
